### Dropping Unneccessary Columns 

```python
fifa.drop(columns=['Photo','Flag','Club Logo'],inplace=True)
```

```python
fifa.shape
```

### Removing symbols and suffixes 

```python
fifa.head()
```

```python
fifa[['Value']]=fifa['Value'].map(lambda x: x.lstrip('€').rstrip('MK'))
```

```python
fifa[['Value']]=fifa['Value'].astype('float')
```

```python
fifa[['Wage']]=fifa['Wage'].map(lambda x: x.lstrip('€').rstrip('K'))
```

```python
fifa[['Wage']]=fifa['Wage'].astype('float')
```

```python
fifa['Joined']=pd.to_datetime(fifa['Joined']).dt.strftime('%Y')
```

```python
fifa[['Contract Valid Until']]=pd.to_datetime(fifa['Contract Valid Until'])
```

```python
fifa[['Height']]=fifa['Height'].map(lambda x: str(x).replace("'",'.'))
```

```python
fifa[['Height']]=fifa['Height'].astype('float')
```

```python
fifa[['Weight']]=fifa['Weight'].map(lambda x: str(x).rstrip('lbs'))
```

```python
fifa[['Weight']]=fifa['Weight'].astype('float')
```

```python
fifa[['Release Clause']]=fifa['Release Clause'].map(lambda x: str(x).lstrip('€').rstrip('MK'))
```

```python
fifa[['Release Clause']]=fifa['Release Clause'].astype('float')
```

```python
fifa[['Joined']]=pd.to_numeric(fifa['Joined'],downcast='integer')
```

```python
fifa.head()
```

```python
fifa.dtypes
```

### Missing Values Imputation

```python
fifa.isnull().sum()
```

```python
round(fifa.isnull().sum()/fifa.shape[0],4)
```

```python
fifa.drop('Loaned From',axis=1,inplace=True)
```

```python
fifa.dropna(thresh=12,inplace=True)
```

```python
fifa[['Release Clause']]=fifa['Release Clause'].fillna(fifa['Release Clause'].mean())
```

```python
fifa[['Joined']]=fifa['Joined'].fillna(fifa['Joined'].mode()[0])
```

```python
fifa[['Club']]=fifa['Club'].fillna(fifa['Club'].mode()[0])
```

```python
fifa[['Contract Valid Until']]=fifa['Contract Valid Until'].fillna(fifa['Contract Valid Until'].mode()[0])
```

```python
fifa[['Position']]=fifa['Position'].fillna(fifa['Position'].mode()[0])
```

```python
fifa[['Jersey Number']]=fifa['Jersey Number'].fillna(fifa['Jersey Number'].mode()[0])
```

```python
fifa.isnull().sum().sum()
```

### 3.	Generate a table containing the top 20 players ranked by Overall score and whose contract expires in 2020.                           
#### a)	What would the average wage for this set of players be?             
#### b)	What is the average age?                          
#### c)	Is there a correlation between the Overall rating and Value for these players?                      


```python
fifa_20=fifa[(fifa['Contract Valid Until']>='2020-01-01') & (fifa['Contract Valid Until']<='2020-12-31')]
```

```python
fifa_20.head(2)
```

```python
fifa_20_overall=fifa_20.sort_values(by=['Overall'],ascending=False).head(20)
fifa_20_overall.head(2)
```

### 3.
### a) What would the average wage for this set of players be?

```python
Average_wage=fifa_20_overall['Wage'].mean()
print(Average_wage,'is the Average wage of top 20 players ranked by Overall score and whose contract expires       in 2020.')
```

#### b) What is the average age?

```python
average_age=fifa_20_overall['Age'].mean()
print(average_age,'Average age of top 20 players ranked by Overall score and whose contract expires in 2020. ')
```

#### c) Is there a correlation between the Overall rating and Value for these players?

```python
sns.heatmap(fifa_20_overall[['Value','Overall']].corr(),annot=True)
```

##### Yes,There is High Positive correlation between Overall rating and Value for these players.
